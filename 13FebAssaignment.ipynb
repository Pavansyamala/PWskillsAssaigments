```python
# Q2)

import inspect

def print_exception_hierarchy(klass, level=0):
    if inspect.isclass(klass) and issubclass(klass, BaseException):
        print('    ' * level + klass.__name__)
        for subclass in klass.__subclasses__():
            print_exception_hierarchy(subclass, level + 1)

# Start with the base class for all exceptions
print("Python Exception Hierarchy:")
print_exception_hierarchy(BaseException)
```
